# **Billboard API**

```python
# Set up billboard API- import dependencies
import billboard
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Find charts with year-end data (omit year_end to get all charts)
# billboard.charts(year_end=True)
```

```python
# Top 100 songs from 2020
chart2020 = billboard.ChartData('hot-100-songs', year=2020, fetch=True)
```

```python
# Get current top 100
current_100 = billboard.ChartData('hot-100')
```

```python
# Filter out the data that you want from the billboard API

# Create lists to get column data
name = [] # name is the column name in Spotify, title is used in Billboard
artists = []
cur_rank = [] # current position in the charts
peak_rank = [] # peak rank in the charts
chrt_weeks = [] # number of weeks on the top 100 chart

# Append each list with the current top 100
for songs in current_100:
    song = current_100[0]
    name.append(songs.title)
    artists.append(songs.artist)
    cur_rank.append(songs.rank)
    peak_rank.append(songs.peakPos)
    chrt_weeks.append(songs.weeks)
```

```python
top_songs = {
    "name": name,
    "artist": artists,
    "current_rank": cur_rank,
    "peak_rank": peak_rank,
    "weeks_on_chart": chrt_weeks
}
top_songs_df = pd.DataFrame(top_songs)
```
